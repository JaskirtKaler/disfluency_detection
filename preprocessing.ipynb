```python
import pandas as pd
```

```python
#Original "SEP-28k-Extended_clips.csv" file with all the extra columns
df_full_extended = pd.read_csv("SEP-28k-Extended_clips.csv")
```

```python
df_full_extended.info()
```

```python
#Dropping unnecessary columns
df_extended = df_full_extended.drop(columns=["Start","Stop","is_probably_host","speaker","clip_silhouette_score","SEP12k","SEP28k-E","Unsure","PoorAudioQuality","DifficultToUnderstand","NaturalPause","Music","NoSpeech"])
```

```python
#Remaining columns (including Train/Test Split columns):
#Show	EpId	ClipId	SEP28k-T	SEP28k-D	Prolongation	Block	SoundRep	WordRep	Interjection	NoStutteredWords
df_extended
```

```python
#map Show EpId ClipID to a filepath column
def toFilePath (row):
    show = str(row["Show"])
    episode = str(row["EpId"])
    clip = str(row["ClipId"])

    return f"SEP-28k_CLIP/{show}/{episode}/{show}_{episode}_{clip}.wav"
```

```python

df_extended["filepath"] = df_extended.apply(toFilePath,axis=1)#apply function row-by-row and create a new filepath column 
```

```python
df_extended.head()
```

```python
df_extended.to_csv("SEP-28k-Filtered_with_Split_and_Path.csv")
```
